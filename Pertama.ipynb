```python
import pandas as pd
import seaborn as sb
import numpy as np
from matplotlib import pyplot as plt
```

### import dataset

```python
# memuat database
df = sb.load_dataset('iris')
print(df.head())
```

### Visualisasi data kontinue

### Histogram

```python
sb.distplot(df['petal_length'], kde=0)
sb.set_style('white')
plt.show()
```

```python
sb.distplot(df['petal_length'], hist=0)
sb.set_style('white')
plt.show()
```

```python
sb.distplot(df['petal_length'])
sb.set_style('white')
plt.show()
```

### Scatter plot 

```python
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df)
sb.set_style('white')
plt.show()
```

### visualisasi data dengan banyak dimensi

```python
sb.set("paper")
sb.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
